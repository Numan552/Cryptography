```python
def vigenere_decrypt(ciphertext, key):
    ciphertext = ciphertext.upper()
    key = key.upper()

    plaintext = ""
    key_length = len(key)
    key_index = 0

    for char in ciphertext:
        if char.isalpha():
            Ci = ord(char) - ord('A')
            Ki = ord(key[key_index % key_length]) - ord('A')

            Pi = (Ci - Ki + 26) % 26
            plaintext += chr(Pi + ord('A'))

            key_index += 1
        else:
            # keep spaces/punctuation unchanged
            plaintext += char

    return plaintext


def main():
    print("Checkpoint 2: Vigenère Cipher Decryption")

    ciphertext = input("Enter the Ciphertext: ")
    key = input("Enter the Key: ")

    if not key.isalpha():
        print("Error: Key must contain only alphabetic characters.")
        return

    plaintext = vigenere_decrypt(ciphertext, key)

    print("\nDecrypted Plaintext:")
    print(plaintext)


if __name__ == "__main__":
    main()
```
