```python
from sympy import factorint, mod_inverse

n = 670726081
d = 12345
#e
factors = factorint(n)
p, q = list(factors.keys())

phi = (p - 1) * (q - 1)
e = mod_inverse(d, phi)

print("Public Key (e):", e)

# Convert text to numbers
def text_to_numbers(text):
    return [ord(c) - 96 for c in text]

# Convert numbers to text
def numbers_to_text(nums):
    result = ""
    for n in nums:
        n = ((n - 1) % 26) + 1
        result += chr(n + 96)
    return result

# Convert ciphertext numbers to text
def ciphertext_to_text(cipher):
    result = ""
    for c in cipher:
        c = c % 26
        if c == 0:
            c = 26
        result += chr(c + 96)
    return result

# Encrypt
def encrypt(data, e, n):
    return [pow(m, e, n) for m in data]

# Decrypt
def decrypt(data, d, n):
    return [pow(c, d, n) for c in data]

# INPUT

plaintext = input("\nEnter text: ")

# input
plaintext = plaintext.replace(" ", "").lower()

# PROCESS

nums = text_to_numbers(plaintext)

# Encrypt
ciphertext = encrypt(nums, e, n)

print("\nCiphertext (first 20 numbers):")
print(ciphertext[:20])

# Ciphertext as text
cipher_text_form = ciphertext_to_text(ciphertext)

print("\nCiphertext (as text - first 200 chars):")
print(cipher_text_form[:200])

# Decrypt
decrypted_nums = decrypt(ciphertext, d, n)
decrypted_text = numbers_to_text(decrypted_nums)

print("\nDecrypted Text (first 200 chars):")
print(decrypted_text[:200])

# VERIFY
print("\nMatch:", decrypted_text == plaintext)
```
