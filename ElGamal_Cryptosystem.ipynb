```python
# Lab 5: Attacking Classic Crypto Systems

def mod_exp(base, exp, mod):
    return pow(base, exp, mod)

def mod_inverse(a, p):
    return pow(a, p - 2, p)

def encode(text):
    return [ord(c) - ord('a') for c in text]

def decode(nums):
    return ''.join(chr(n + ord('a')) for n in nums)

# Encryption
def encrypt(p, g, y, k, plaintext):
    encoded = encode(plaintext)
    c1 = mod_exp(g, k, p)

    ciphertext = []
    for m in encoded:
        c2 = (m * mod_exp(y, k, p)) % p
        ciphertext.append((c1, c2))

    return ciphertext

# Decryption
def decrypt(p, x, ciphertext):
    decrypted_nums = []

    for c1, c2 in ciphertext:
        s = mod_exp(c1, x, p)
        s_inv = mod_inverse(s, p)
        m = (c2 * s_inv) % p
        decrypted_nums.append(m)

    return decode(decrypted_nums)

p = 593
g = 3
x = 89
k = 41

# Task 1: Compute public key
y = mod_exp(g, x, p)

print("Public Key (p, g, y):", (p, g, y))
print("Private Key x:", x)

# Input
plaintext = input("Enter plaintext (lowercase only): ").lower()
plaintext = ''.join([c for c in plaintext if c.isalpha()])

# Encryption
ciphertext = encrypt(p, g, y, k, plaintext)

print("\nCiphertext (first 10 pairs):")
for pair in ciphertext[:10]:
    print(pair)
print("...")

# Decryption
decrypted_text = decrypt(p, x, ciphertext)

print("\nDecrypted Text:")
print(decrypted_text)

# Verification
if decrypted_text == plaintext:
    print("\n Decryption successful! Matches original plaintext.")
else:
    print("\n Decryption failed!")
```
